```python
features = df.columns
for feature in features:
    plt.hist(df[feature])  
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(f'Histogram of {feature}')
    plt.show()
```

```python
features = df.columns
for feature in features:
    plt.figure()
    plt.boxplot(df[feature])  
    plt.xlabel("Features")
    plt.ylabel("Count")
    plt.title(f'Boxplot of {feature}')
    plt.show()
    
```

```python
print(df.dtypes)
df['SepalLengthCm'] = pd.to_numeric(df['SepalLengthCm'], errors='coerce')
df['SepalWidthCm'] = pd.to_numeric(df['SepalWidthCm'], errors='coerce')
df = df.dropna(subset=['SepalLengthCm', 'SepalWidthCm'])
result = df['SepalLengthCm'] / df['SepalWidthCm']
print(result)
```
